```python
import pandas as pd
import numpy as np

brands = pd.read_csv("brands.csv") 
finance = pd.read_csv("finance.csv")
info = pd.read_csv("info.csv")
reviews = pd.read_csv("reviews.csv")

# Start coding here...
```

# Instructions 

## 1. What is the volume of products and average revenue for Adidas and Nike products based on listing price quartiles?

    - Label products priced up to quartile one as "Budget", quartile two as "Average", quartile three as "Expensive", and quartile four as "Elite".


```python
# 1 merge DataFrames
merged = brands.merge(finance, on='product_id').merge(info, on='product_id').merge(reviews, on='product_id')
merged = merged.dropna()
# 2 create a price_label column with the price quartiles cuts
merged['price_label'] = pd.qcut(merged["listing_price"], q=4, labels=["Budget", "Average", "Expensive", "Elite"])
```

    - Store as a pandas DataFrame called adidas_vs_nike containing the following columns: "brand", "price_label", "num_products", and "mean_revenue". All numeric values should be rounded to two decimal places.


```python
# 3 group by brand and price_label and calculate the number of products and mean revenue
adidas_vs_nike = merged.groupby(['brand', 'price_label'], as_index = False , observed=True).agg(
    num_products = ('price_label','count'),
    mean_revenue =('revenue','mean')
).round(2)
adidas_vs_nike
```

## 2. Do any differences exist between the word count of a product's description and its mean rating?

    - Split product description length into bins of 100 characters and calculate the average rating and number of reviews.

```python
# 4 calculate the length of each product's description
merged['description_length'] = merged['description'].str.len()
# 5 group the description lengths into bins of 100 characters
lengthes = [0, 100, 200, 300, 400, 500, 600, 700]
labels = ["100", "200", "300", "400", "500", "600", "700"]
merged["description_length"] = pd.cut(merged["description_length"], bins=lengthes, labels=labels)

```

    - Store the results as a pandas DataFrame called description_lengths containing the following columns: "description_length", "mean_rating", "num_reviews", again rounding numeric values to two decimal places.



```python
# Group by the bins
description_lengths = merged.groupby("description_length", as_index=False, observed=True).agg(
    mean_rating=("rating", "mean"), 
    num_reviews=("reviews", "count")
).round(2)

description_lengths
```
