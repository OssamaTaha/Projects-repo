Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)? Save as a string variable called peak_night_crime_location.

Time on The dataframe is in 24h format, and is not human friendly cuz it does not have the ":" separating hours and mins.
Lets fix it


```python
# Function to convert TIME OCC column to datetime.time
from datetime import datetime, time

def convert_time_occ_column(crimes):
    crimes['TIME OCC'] = crimes['TIME OCC'].apply(lambda x: datetime.strptime(f'{x:04}', '%H%M').time())
    print(f"Column 'TIME OCC' converted to type: {crimes['TIME OCC'].dtype}")

# Convert TIME OCC column
convert_time_occ_column(crimes)
```

```python
# Define the nighttime range
night_start = time(22, 0)  # 10:00 PM
night_end = time(3, 59)    # 3:59 AM

# Function to filter nighttime crimes
def is_nighttime(crime_time):
    return (crime_time >= night_start) or (crime_time <= night_end)

# Filter for nighttime crimes
nighttime_crimes = crimes[crimes['TIME OCC'].apply(is_nighttime)]

print("\nFiltered DataFrame with nighttime crimes:")
nighttime_crimes.head()
```

## Which area has the largest frequency of night crimes

```python
# Find the area with the largest frequency of nighttime crimes
most_frequent_area = nighttime_crimes['AREA NAME'].value_counts().idxmax()
frequency = nighttime_crimes['AREA NAME'].value_counts().max()

print(f"The area with the largest frequency of nighttime crimes is: {most_frequent_area} with {frequency} crimes")
```
